```python
s = pd.Series([1, 2, 3, 4, 5, 5, 88, np.nan, 8, 0, 77])
# isnull  (series + df)
print(s.isnull())
# notnull (series + df)
print(s.notnull())
# hasnans (series)
print(s.hasnans)
```

```python
# dropna (series + df)
print(s.dropna())

# dropna df
df = pd.DataFrame([['abc', 'a', '90'], 
                 ['xyz', 'c', '70'], 
                 ['pqr', 'b', '10'], 
                 ['lmn', 'a', '80'],
                 [np.nan, np.nan, np.nan], 
                 ['abc', 'a', np.nan],
                [np.nan, 'a', np.nan]], columns=['Name', 'Grade', 'Sal'])
print(df)
```

```python
# dropna df

# drops row even if one column is nan
print(df.dropna(how='any'))
# print(df.dropna())
```

```python
# drops only full nan row
print(df.dropna(how='all'))
```

```python
# drop row based on specific column
print(df.dropna(subset=['Name', 'Grade']))
```

```python
# apply on df
points_df = pd.DataFrame(
{
    '1st point': [(3,4), (-6,5), (0,0), (-13,4), (5,4), ],
    '2nd point': [(-3,4), (0,0), (2,2), (9,10), (1,1), ],
})

points_df
```

```python
def euclidean(row):
    pta = row['1st point']
    ptb = row['2nd point']
    
    return ((pta[0] + ptb[1])**2 + (pta[1] + ptb[0])**2)**0.5

points_df['Distance'] = points_df.apply(euclidean, axis=1)
```
